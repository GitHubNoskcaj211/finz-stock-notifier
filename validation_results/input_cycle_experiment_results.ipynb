```python
from matplotlib import pyplot as plt
import pandas as pd
```

This experiment ranges the input cycle from 14 days to 350 days in 14 day increments. The start day of the cycle is always the last day of the cycle. This test represents what investing is like if you have no money saved to invest with at the beginning and use different investment schedules. Only DCA and lump sum methods are used and fractional shares are always used to represent an expected value. This result shows that the shorter your input cycle (the earlier you put money into the market), the higher your gains.

All the experiments are paired, so the mean is a fair comparison method.

Results show the more frequent you put money into the market (shorter investment cycles) the higher your returns. The disparity is greater for the DCA model than the lump sum model (annual roi drops just under 1% for DCA whereas about 0.5% for lump sum).

```python
file_name = 'input_cycle_experiment.csv'
```

```python
df = pd.read_csv(file_name)
```

```python
def plot_metric_versus_input_cycle(df, model_name, metric_name):
    grouped_data = df.groupby('investment_input_cycle_days')[f'{model_name}_{metric_name}'].agg(['mean', 'std']).reset_index()
    fig, ax = plt.subplots()
    ax.errorbar(grouped_data['investment_input_cycle_days'], grouped_data['mean'], fmt='o', capsize=5, label='MeanMean')
    ax.set_xlabel('investment_input_cycle_days')
    ax.set_ylabel('Metric Values')
    ax.set_title(f'Model {model_name} | Metric {metric_name} | Metric Value vs. Investment Cycle')
    ax.legend()
    return fig
```

```python
plot_metric_versus_input_cycle(df, 'constant_dollar_random_model', 'total_annual_roi')
plot_metric_versus_input_cycle(df, 'lump_sum_model', 'total_annual_roi')
pass
```

```python
plot_metric_versus_input_cycle(df, 'constant_dollar_random_model', 'total_roi')
plot_metric_versus_input_cycle(df, 'lump_sum_model', 'total_roi')
pass
```

```python
plot_metric_versus_input_cycle(df, 'constant_dollar_random_model', 'average_price')
plot_metric_versus_input_cycle(df, 'lump_sum_model', 'average_price')
pass
```
